# The Independent Samples t Test compares the means of two independent groups in order to determine whether there is statistical evidence that the associated population means are significantly different. The Independent Samples t Test is a parametric test.

# Hypothesis Testing : Comparison between two different groups

# 2 Sample Independent t-test

# Test whether mean Age of Manual cars and Auto cars is same 

- H0: No Significant Difference in Mean Age of two Groups
- H1: Significant Difference in Mean Age of two Groups

```python
#Observe the mean prices of two groups

data[['Age','AutoType']].groupby(['AutoType']).mean().reset_index()
```

```python
manual = data[data['AutoType']=='Manual']
auto = data[data['AutoType']=='Auto']
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(auto.Age, manual.Age)
```

# alpha = 0.05, statistic = mean

# As p-value greater than 0.05, Failed to reject Null Hypothesis (H0)
